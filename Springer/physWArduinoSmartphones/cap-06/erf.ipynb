# The error function

The error function provides a way to estimate the cumulative of a gaussian. It is defined in ``scipy.special`` package. With the code below we plot the pdf and the cdf of a normal distribution, together with the erf(x) in green.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
from scipy.special import erf

x = np.arange(-5, 5, .01)
y = norm(0., 1.)
plt.plot(x, y.pdf(x), '-')
plt.plot(x, y.cdf(x), '-')
plt.plot(x, erf(x), '-')
plt.show()
```
