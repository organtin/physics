```python
import numpy as np
import matplotlib.pyplot as plt

def lin(x, A = 1, B = 1):
    return A*x+B

def f(x):
    return np.sin(x)

def g(x):
    return (x-3)**2

x = np.arange(0, 2*np.pi, .5)

from scipy.stats import pearsonr
corr, pv = pearsonr(x, lin(x))

plt.plot(x, lin(x), '-')
plt.plot(x, lin(x), 'o')
plt.annotate('$r_{{x,y}}={}$'.format(corr), xy = (.1,.9), xycoords = 'axes fraction')
plt.annotate('$p$-value={}'.format(pv), xy = (.1,.8), xycoords = 'axes fraction')
sx = np.std(x)
sy = np.std(lin(x))
COV = np.corrcoef(x, lin(x))
print(COV)
COV = np.cov(x, lin(x))
print(COV[0][1]/np.sqrt(COV[0][0]*COV[1][1]))
```

```python
corr, pval = pearsonr(x, f(x))

plt.figure(figsize=(8.5,4))
plt.plot(x, f(x), '-')
plt.plot(x, f(x), 'o')
plt.annotate('$r_{{x,y}}={}$'.format(corr), xy = (.5,.9), xycoords = 'axes fraction')
plt.savefig('pcc-sine-2pi')
COV = np.corrcoef(x, f(x))
print(COV)
COV = np.cov(x, f(x))
print(COV[0][1]/np.sqrt(COV[0][0]*COV[1][1]))
```

```python
corr, pval = pearsonr(x, g(x))

plt.figure(figsize=(8.5,4))
plt.plot(x, g(x), '-')
plt.plot(x, g(x), 'o')
plt.annotate('$r_{{x,y}}={}$'.format(corr), xy = (.1,.9), xycoords = 'axes fraction')
plt.annotate('$p$-value={}'.format(pval), xy = (.3,.9), xycoords = 'axes fraction')
plt.savefig('pcc-parabola.png')
COV = np.corrcoef(x, g(x))
print(COV)
COV = np.cov(x, g(x))
print(COV[0][1]/np.sqrt(COV[0][0]*COV[1][1]))
```
