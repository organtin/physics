```python
import numpy as np
import matplotlib.pyplot as plt

def y(x):
    return np.sin(x)

def P(y, C):
    return C/(np.pi*np.sqrt(1-y*y))

x = np.arange(0, 70, .1)
plt.plot(x, y(x))
plt.show()
```

```python
plt.figure(figsize=(8.5,4))
h, b, r = plt.hist(y(x), bins = 30)
C = sum(h)*(b[1]-b[0])
xr = np.arange(-1,1,.01)
plt.
plt.plot(xr, P(xr, C), '-')
```
